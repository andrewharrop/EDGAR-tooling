```python
from XBRLField import XBRLField
```

```python
AAPL=XBRLField('AAPL')
```

```python
netIncomeLoss =AAPL.fetchField("NetIncomeLoss")
```

```python
netIncomeLoss
```

```python
allAnnual = netIncomeLoss["annual"]
allAnnual = allAnnual[~allAnnual["frame"].isna()]
allAnnual = allAnnual[~allAnnual["frame"].str.contains("Q")]
allAnnual = allAnnual[["val", "frame"]]
allAnnual['frame'] = allAnnual['frame'].str[2:]
allAnnual.rename(columns={'frame':'year'}, inplace=True)
allAnnual.reset_index(drop=True, inplace=True)
```

```python
allAnnual
```

```python
# Normalize
allAnnual["val"] = allAnnual["val"].astype(float) / allAnnual["val"].iloc[0]
```
